```python
#import data from github

import pandas as pd
url = 'https://github.com/Syukrondzeko/Fellowship/blob/main/Seasons_Stats.csv?raw=true'

data = pd.read_csv(url,index_col=0)

data.head(5)
```

## Data Pre-processing

### Selecting data only in 2017

```python
#only data in 2017 will be used
df= data[data.Year == 2017]
df= df.reset_index(drop=True)
df.head()
```

### Dealing with duplicates values

```python
#check Player data
df.Player.value_counts()
```

```python
#example of the data that have multiple values
df[df.Player == "Ersan Ilyasova"]
```

```python
#drop duplicates values and keep the first records from duplicates
#with an assumption that the first row contains the latest data 

df = df.drop_duplicates("Player", keep='first')
df= df.reset_index(drop=True)
```

### Dealing with missing values

droping columns that have missing values in their entire data row

```python
#check missing values percentage in every columns
df.isnull().sum() / df.shape[0] * 100
```

```python
#drop columns that have 100% missing vales
NBA_df = df.drop(['blanl', 'blank2'], axis=1)
NBA_df.head()
```

```python
#replace missing values in the data with value of 0.0 as the null total percentage relatively small
#so, replaced by 0.0 would not arise any issues.
NBA_df = NBA_df.fillna(0.0)
```

## Analysing data

### 1.	Who is the youngest and oldest player in the NBA in 2017 for each team (Tm) ?

```python
#get only the required columns (Team, PLayer Name,Age)
Q1_df = NBA_df[['Tm','Player', 'Age']]
Q1_df
```

```python
#create a Youngest dataframe by grouping the player by team and search the minimum values of the age in each team
Youngest = Q1_df[Q1_df.groupby('Tm')['Age'].transform('min') == df['Age']]
Youngest = Youngest.sort_values(by=['Tm']).reset_index(drop=True)
Youngest
```

```python
#create an Oldest dataframe by grouping the player by team and search the maximum values of the age in each team
Oldest = Q1_df[Q1_df.groupby('Tm')['Age'].transform('max') == df['Age']]
Oldest = Oldest.sort_values(by=['Tm']).reset_index(drop=True)
Oldest
```

### 2.	Which player has the most minutes played (MP) in each position (Pos)? 

```python
#get only the required columns (Pos, PLayer Name, MP)
Q2_df = NBA_df[['Pos','Player', 'MP']]
Q2_df
```

```python
#create a MostMP_per_Pos dataframe by grouping the player by position and search the maximum values of the 'MP' in each position
MostMP_per_Pos = Q2_df[Q2_df.groupby('Pos')['MP'].transform('max') == df['MP']]
MostMP_per_Pos = MostMP_per_Pos.reset_index(drop=True)
MostMP_per_Pos
```

### 3.	Which team has the highest average total rebound percentage (TRB%), assist percentage (AST%), steal percentage (STL%), and block percentage (BLK%)? 

```python
#get only the required columns (Pos, PLayer Name, MP)
Q3_df = NBA_df[['Tm','TRB%','AST%','STL%','BLK%']]
Q3_df
```

```python
#get the average of total variables and grouped by team
total_avg = Q3_df.groupby(['Tm']).mean()
total_avg
```

Highest Average of Total Rebound Percentage(TRB%)

```python
#find the max values of total average in TRB%
Highest_avg_TRB = total_avg[total_avg['TRB%'] == total_avg['TRB%'].max()]
Highest_avg_TRB = Highest_avg_TRB[['TRB%']]
Highest_avg_TRB
```

Highest Average of Total Assist Percentage(AST%)

```python
#find the max values of total average in AST%
Highest_avg_AST = total_avg[total_avg['AST%'] == total_avg['AST%'].max()]
Highest_avg_AST = Highest_avg_AST[['AST%']]
Highest_avg_AST
```

Highest Average of Total Steal Percentage(STL%)

```python
#find the max values of total average in STL%
Highest_avg_STL = total_avg[total_avg['STL%'] == total_avg['STL%'].max()]
Highest_avg_STL = Highest_avg_STL[['STL%']]
Highest_avg_STL
```

Highest Average of Total Block Percentage(BL%K)

```python
#find the max values of total average in BLK%
Highest_avg_BLK = total_avg[total_avg['BLK%'] == total_avg['BLK%'].max()]
Highest_avg_BLK = Highest_avg_BLK[['BLK%']]
Highest_avg_BLK
```

### 4.	Who is the best player in your opinion based on his record stats? note: you can refer to variables point (PTS), assists, rebounds, or anything else. A combination of several variables would be nice. 

The best player would be the one that has the highest overall rating in average. To determine the best player in NBA 2017, I will look at several variables. These variables are:
- PER (Player efficiency rating)
- eFG% (Effective Field Goal Percentage)
- TS% (True Shooting Percentage)
- ORB% (Offensive Rebound Percentage)
- DRB%( Defensive Rebound Percentage)
- TRB (Total Rebounds)
- AST% (Assist Percentage)
- STL% (Steal Percentage)
- BLK% (Block Percentage)
- PTS (Points)
- BPM (Box Plus/Minus)
- WS (Win Shares)
- WS/48 (Win Shares)

These were chosen by research on how to measure the NBA player.
Reference: 
- https://benchboss.ai/efficiency-the-most-important-stat-in-basketball/
- https://bleacherreport.com/articles/1813902-advanced-nba-stats-for-dummies-how-to-understand-the-new-hoops-math

```python
#get only the required columns (Pos, PLayer Name, MP)
Q4_df = NBA_df[['Player','Pos','Tm','PER','eFG%','TS%','ORB%','DRB%','TRB','AST%','STL%','BLK%','PTS','BPM','WS','WS/48']]
Q4_df
```

```python
#create a column that contain the mean of each variable
Q4_df['avg_overall_metrics'] = Q4_df.mean(axis=1)
Q4_df
```

Best Player NBA 2017

```python
#find the player that has the highest 'avg_overall_metrics'
Best_Player = Q4_df[Q4_df['avg_overall_metrics']== Q4_df['avg_overall_metrics'].max()]
Best_Player
```
