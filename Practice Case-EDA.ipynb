```python
#replace the datatype into integer
data['attack_passes'] = data['attack_passes'].astype(int)
data['attack_passes_long'] = data['attack_passes_long'].astype(int)
data['attack_passes_back'] = data['attack_passes_back'].astype(int)
data['defence_clearances'] = data['defence_clearances'].astype(int)
```

## Q3. Which team has the best attack?

In football(soccer), team that has the best attack usually use an offensive play/stategy. This team is usually aiming for scoring many points or goals. This team is also controlling game levels by maintaining the ball possession. However, I would like to include the counter attack variable too altough this is used by the teams that have defensive stategy.

Therefore, the variables that are suitable to find the team that has the best attack are:
- attack_scored
- attack_passes
- attack_passes_through
- attack_passes_long
- attack_passes_back
- attack_crosses
- attack_corners_taken
- attack_shots
- attack_shots_on_target
- attack_goals_headed
- attack_goals_box
- attack_goals_outsidebox
- attack_goals_counter
- attack_posession
- attack_pass_accuracy

```python
#select the required columns
attack = data[['attack_scored','attack_passes','attack_passes_through',
                   'attack_passes_long','attack_passes_back','attack_crosses',
                   'attack_corners_taken','attack_shots','attack_shots_on_target',
                   'attack_goals_headed','attack_goals_box','attack_goals_outsidebox',
                   'attack_goals_counter','attack_posession','attack_pass_accuracy']]
attack
```

I would like to include the league position, total win, draw, and lose of the team but later after calculatin the metric. So, it will not impact calculationf for attacking metric as those variable does not determine wheter the team has best attack or not.

```python
#create the new column of attacking metric based on the their mean
#based on group

attack['attack_metric'] = attack.mean(axis=1)
attack
```

```python
#create club data and join it with the attack data
club_info = data[['Team','general_league_position','general_won',
                  'general_draw','general_lost']]
attack_team = club_info.join(attack)
attack_team
```

```python
#get the highest attack metric
Best_attack_team = attack_team[attack_team['attack_metric']== attack_team ['attack_metric'].max()]
Best_attack_team
```

```python
Best_attack_team[['Team','attack_metric','general_league_position','general_won',
                  'general_draw','general_lost','attack_scored']]
```

After cross-check who was the best attaking team in Premier League 2018/2019, it is true that the team was Manchester City.

## Q4. Which team has the best defense? 

In contrast with best attack team before, team that has the best defense should be measured by how well the team bloking the opponent to score, saving their goal and defending their field by wining the ball.

## Q5. Which team is good in the financial aspect?  

Get the coloumns that related to the club's financial information.
There would be 4 things that I do. 
- First, team that has good in financial overall aspect.
- Second, team that has good financial in tv revenue.
- Third, team that has good financial in team market.
- Lastly, team that has good financial in market average.

```python
#find the team with highest 'tv_fin'
tv_fin = finance[finance['finance _tv_revenue']== finance['finance _tv_revenue'].max()]
tv_fin
```

### Team that has good financial in team merket.

```python
#find the team with highest 'team_market_fin'
team_market_fin = finance[finance['finance _team_market']== finance['finance _team_market'].max()]
team_market_fin
```

### Team that has good financial in market average.

```python
#find the team with highest 'market_avg_fin'
market_avg_fin = finance[finance['finance _market_average']== finance['finance _market_average'].max()]
market_avg_fin
```
