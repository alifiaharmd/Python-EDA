## 2015

```python
df2015.drop(['Region','Standard Error', 'Dystopia Residual'],axis=1,inplace=True)
df2015.columns = ["Country", "Rank", "Happiness_score", "GDP_per_capita", "Family",
"Life_expectancy", "Freedom", "Trust", "Generosity"]
df2015["Year"] = 2015
df2015.head()
```

## 2016

```python
df2016.drop(['Region','Lower Confidence Interval', 'Upper Confidence Interval', 'Dystopia Residual'],axis=1,inplace=True)
df2016.columns = ["Country", "Rank", "Happiness_score", "GDP_per_capita", "Family",
"Life_expectancy", "Freedom", "Trust", "Generosity"]
df2016["Year"] = 2016
df2016.head()
```

## 2017

```python
df2017.drop(['Whisker.high', 'Whisker.low', 'Dystopia.Residual'],axis=1,inplace=True)
df2017.columns = ["Country", "Rank", "Happiness_score", "GDP_per_capita", "Family",
"Life_expectancy", "Freedom", "Generosity", "Trust"]
df2017["Year"] = 2017
df2017.head()
```

## 2018

```python
df2018.columns = ["Rank", "Country", "Happiness_score", "GDP_per_capita", "Family",
"Life_expectancy", "Freedom", "Generosity", "Trust"]
df2018["Year"] = 2018
df2018.head()
```

## 2019

```python
df2019.columns = ["Rank", "Country", "Happiness_score", "GDP_per_capita", "Family",
"Life_expectancy", "Freedom", "Generosity", "Trust"]
df2019["Year"] = 2019
df2019.head()
```

### Create binning for Happiness Rank

```python
bin_labels = ['Top', 'Top-Mid', 'Mid', 'Low-Mid', 'Low']

def bin(dfyear):
    dfyear['Class'] = pd.qcut(dfyear['Rank'], q=5, labels=bin_labels)
    return dfyear

df2015 = bin(df2015)
df2016 = bin(df2016)
df2017 = bin(df2017)
df2018 = bin(df2018)
df2019 = bin(df2019)
```

```python
df2015.head()
```

### Reorder the columns

```python
def reorder(df_year):
    df_year=df_year[["Country", "Year", "Happiness_score", "GDP_per_capita", "Family", 
                   "Life_expectancy", "Freedom", "Generosity", "Trust", "Class", "Rank"]]
    return df_year

df2015 = reorder(df2015)
df2016 = reorder(df2016)
df2017 = reorder(df2017)
df2018 = reorder(df2018)
df2019 = reorder(df2019)
```

### Append dataframe

```python
df = df2015
df = df.append(df2016, ignore_index = True)
df = df.append(df2017, ignore_index = True)
df = df.append(df2018, ignore_index = True)
df = df.append(df2019, ignore_index = True)
df.head()
```

### Set new indexes

```python
df = df.set_index(['Country', 'Year', 'Rank'])
df.head()
```
